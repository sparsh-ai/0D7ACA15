### Import Libraries

```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

### Open Image

```python
tst_img = cv2.imread('images/eye_face.jpg')
tst_img = cv2.cvtColor(tst_img, cv2.COLOR_BGR2RGB)
plt.imshow(tst_img)
```

### Create Classifier

```python
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
```

### It's a Kind of Magic

```python
def detect_eyes(fix_img):
    eyes_rects = eye_classifier.detectMultiScale(fix_img)
    
    for (x, y, w, h) in eyes_rects:
        cv2.rectangle(fix_img,
                     (x, y),
                     (x+w, y+h),
                     (255, 255, 255),
                     10)
        
    return fix_img
```

### Show Results

```python
result = detect_eyes(tst_img)
plt.imshow(result);
```
